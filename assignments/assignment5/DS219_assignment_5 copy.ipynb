### Hypothesis Formulation

#### 1. Null: There is no correlation between the quality of red wines and its fixed acidicity./ Alt: There is a significant correlation.
#### 2. Null: The density of red wine is the same as the density of white wine. / Alt: There is difference between them.

### Hypothesis Testing/Test the formulated hypotheses using statistical tests or visualizations

```python
# 1st Hypothesis/Pearson 
alpha=0.05
cor_coef, pvalue =  stats.pearsonr(df_red["quality"], df_red["fixed acidity"])
print(f"Correlation Coefficient: {cor_coef}")
print(f"P-value: {pvalue}")
alpha = 0.05
if pvalue > alpha:
    print("Fail to reject the null hypothesis. There is no significant correlation.")
else:
    print("Reject the null hypothesis. There is a significant correlation.")
```

```python
#2nd Hypothesis 
alpha=0.05
t_stat, pvalue =stats.ttest_ind(df_red["density"], df_white["density"])
print(f"T-statistic: {t_stat}")
print(f"P-value: {pvalue}")
if pvalue < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
